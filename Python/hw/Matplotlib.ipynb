3. Постройте графики еще нескольких элементарных функций (на том же рисунке) и сделайте к ним подписи (легенду) с формулами (в подписях можно использовать LaTex, как обычно, в знак '$'). Добавьте какие-нибудь подписи к осям координат и название графика. Сохраните изображение в файл одной строчкой на Python.
***

```python
import matplotlib.pyplot as plt
import numpy as np

X = np.linspace(-5,5, 50)

Y = [0.5*x*x + 1 for x in X]
plt.plot(X, Y, 'b', label='$y=0.5x^2 + 1$')

y = [np.cos(x) for x in X]
plt.plot(X,y, label ='$y=cos(x)$')

y = [x*2 + 1 for x in X]
plt.plot(X,y, label = '$y=2x + 1$')

plt.legend()
plt.xlabel('$X$')
plt.ylabel('$Y$')
plt.title("Functions")
plt.savefig('functions.png')
```

4. Изучите документацию matplotlib и попробуйте построить на одном изображении 4 системы координат, на верхних двух - квадратичную и кубическую параболы, на нижних - экспоненту и логарифм.
***

```python
x = np.linspace(-10,10, 20)

plt.subplot(221)
plt.plot(x,x**2)
plt.title('$x^2$')

plt.subplot(222)
plt.plot(x,x**3,'g')
plt.title('$x^3$')

plt.subplot(223)
plt.plot(x,np.exp(x))
plt.title('exp(x)')

plt.subplot(224)
plt.plot(x,np.log(x))
plt.title('log(x)')
```
