```python
import numpy as np
import scipy
import matplotlib.pyplot as plt
from scipy.optimize import minimize
%matplotlib inline
```

```python
x = [2, 4, 6, 7, 10]
y = [1, 3, 5]
plt.plot(x, [y for y in range(5)], 'bx-')
```

```python
x = [np.random.uniform(-2,3) for i in range(500)]
x = np.linspace(-2,3, 500)
y = [0.5*j + 1 + np.random.normal(0, 0.2) for j in x]
plt.plot(x, y, 'gx')
```

```python
def linear(coeffs, x_i): 
    return coeffs[0]*x_i + coeffs[1]

def loss(coeffs):
    predictions = np.array([linear(coeffs, x_i) for x_i in x])
    ans = np.array(y)
    return sum((predictions - ans)**2)

plt.plot(x, y, 'gx')    
#plt.plot(x, loss([0.5,1], 'gx')  
plt.plot(x, [linear([0.5,1],i) for i in x], 'b')
```

```python
from scipy.optimize import minimize
res =  minimize(loss, [1,10]).x
res
```

```python
x = [np.random.uniform(-2,3) for i in range(500)]
x = np.linspace(-2,3, 500)
y = [0.5*j + 1 + np.random.normal(0, 0.2) for j in x]
plt.plot(x, y, 'gx')

x1 = [np.random.uniform(-2,3) for i in range(200)]
y1 = [3+np.random.normal(0, 0.2) for _ in range(200)]
plt.plot(x, y, 'bx')

xc = np.concatenate([x,x1])
yc = np.concatenate([y,y1])

def linear(coeffs, x_i): 
    return coeffs[0]*x_i + coeffs[1]

def loss(coeffs):
    predictions = np.array([linear(coeffs, x_i) for x_i in xc])
    ans = np.array(yc)
    return sum((predictions - ans)**2)

res =  minimize(loss, [1,10]).x

plt.plot(xc, yc, 'gx')    
#plt.plot(x, loss([0.5,1], 'gx')  
plt.plot(xc, [linear(res,i) for i in xc], 'b')

def loss(coeffs):
    predictions = np.array([linear(coeffs, x_i) for x_i in xc])
    ans = np.array(yc)
    return sum(abs(predictions - ans))

res =  minimize(loss, [1,10]).x

plt.plot(xc, [linear(res,i) for i in xc], 'r')


```
